# Part 1A: Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import csv

# load all the cvs data files
crash_data = pd.read_csv('Traffic_Crashes_-_Crashes.csv')
vehicle_data = pd.read_csv('Traffic_Crashes_-_Vehicles.csv')

crash_data = crash_data.sort_values(by=['RD_NO'])
vehicle_data = vehicle_data.sort_values(by=['RD_NO'])

df = pd.merge(crash_data,vehicle_data, on = 'RD_NO')
```

```python
# Globally set options to display all columns and display all rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

We used the following part of the code to analyze different features and what parts we should have used in the model (it has been done for the numerous columns and Vehicle type column provided as an example)

```python
df.groupby('VEHICLE_TYPE').count()
```

Here we clean the data to inculde only on-road vechicles and passenger vehicle types 
and excluding unknown values for the traffic type, lightning condition and weather columns 

```python
df = df[df.UNIT_TYPE == 'DRIVER']
df = df[df.TRAFFICWAY_TYPE != 'UKNOWN']
df = df[df.WEATHER_CONDITION != 'UKNOWN']
df = df[df.VEHICLE_TYPE == 'PASSENGER']
df = df[df.LIGHTING_CONDITION != 'UNKNOWN']
df = df[df.PRIM_CONTRIBUTORY_CAUSE != 'NOT APPLICABLE']
df.head()
```

# Creation of new DataFrame with the features we deemed more important

The new dataframe is called 'df_new', and only consists of selected features. The existing dataframe
'df' has all columns.

```python
df_new = df.loc[:,['RD_NO','WEATHER_CONDITION','INJURIES_TOTAL','DAMAGE','CRASH_DAY_OF_WEEK','CRASH_MONTH',\
 'CRASH_HOUR','LIGHTING_CONDITION','MANEUVER','TRAFFICWAY_TYPE','PRIM_CONTRIBUTORY_CAUSE',\
 'POSTED_SPEED_LIMIT','DEVICE_CONDITION','TRAFFIC_CONTROL_DEVICE','ALIGNMENT','ROADWAY_SURFACE_COND',\
 'LATITUDE','LONGITUDE','LOCATION']]
```

```python
# Checking whether there are Null values in the columns
NullsPerColumn = pd.DataFrame((df_new.isnull().sum())/len(df_new.index))
NullsPerColumn.columns=['Percentage Null']
NullsPerColumn
```

```python
# removing any remaining NaN values
df_new = df_new.dropna(axis=0)
```

We note that the column 'NUM_PASSENGERS' does not include the driver of the vehicle. Hence, we check
the following:
If the 'NUM_UNITS' = 1, this implies that there was only one driver involved in the crash, and we +1 to
the total number of passengers in the vehicle.
If the 'NUM_UNITS' > 1, this implies that there was more than one vehicle involved in the crash, and we
add that value to the total number of passengers in the vehicle.

```python
df.NUM_PASSENGERS = df.NUM_PASSENGERS.apply(lambda x: 1 if np.isnan(x) else x+1)
df['NUM_DRIVERS'] = df.NUM_UNITS.apply(lambda x: x if x==1 else x+1) 
df.NUM_PASSENGERS = df.NUM_PASSENGERS + df.NUM_DRIVERS
df.NUM_PASSENGERS.unique()
```

Next we find number of injuries per person involved in the accident by deviding the total number of the ing\juries in the accident by a total number of people involved

```python
df_new['INJURIES_PER_PASSENGER'] = df_new['INJURIES_TOTAL']/ df['NUM_PASSENGERS']
```

```python
import matplotlib.pyplot as plt
plt.hist(df_new['INJURIES_PER_PASSENGER'], bins = 50)
plt.show()
print(df_new['INJURIES_PER_PASSENGER'].unique())
```

Assigning danger score according to the number of injuries per person in the vehicle
We choose to assign 4 danger score ratings to account for all accidents depending on the amount of injuries

```python
df_new['DANGER_SCORE'] = df_new.INJURIES_PER_PASSENGER.apply(lambda x: 4 if x>2 else x)
df_new.DANGER_SCORE = df_new.DANGER_SCORE.apply(lambda x: 3 if (x>1 and x<=2) else x)
df_new.DANGER_SCORE = df_new.DANGER_SCORE.apply(lambda x: 2 if (x>0 and x<=1) else x)
df_new.DANGER_SCORE = df_new.DANGER_SCORE.apply(lambda x: 1 if x==0 else x)
df_new.groupby('DANGER_SCORE').count()
```

In computing the danger score, we intend to include the physical damage of the vehicle. First, we check the
unique values of the 'DAMAGE' column, then decide what weight to add to it in the computation of the
danger score.

```python
df_new.DAMAGE.unique()
```

There are 3 unique values (' 501 - 1,500' ), so we assign values of 1, 2 and 3
to these respectively, then add a weight of 'w' in the multiplication.

```python
df_new.DAMAGE = df_new.DAMAGE.apply(lambda x: 1 if x == '$500 OR LESS' else (2 if x == '$501 - $1,500' else 3))
df_new.DAMAGE.unique()
```

# Create a copy of the dataframe to test the weights

```python
df_new.DANGER_SCORE.unique()
```

```python
w = 0.5 # assign a weight of 0.5 (can be changed)
df_new['COMBINED_DANGER_SCORE'] = df_new.DANGER_SCORE*(w*df_new.DAMAGE)
```

```python
df_new.COMBINED_DANGER_SCORE.unique()
```

With a weight of 0.5, we have 8 different scores to bin into 4 categories again. X = 1, if current score is 0.5
and 1 X = 2, if current score is 1.5 and 2 X = 3, if current score is 3 and 4 X = 4, if current score is 4.5 and 6
Now we check the count for each value of 'COMBINED_DANGER_SCORE'

```python
df_new.COMBINED_DANGER_SCORE.value_counts()
```

We would need to bin the COMBINED_DANGER_SCORE again

```python
df_new.COMBINED_DANGER_SCORE = df_new.COMBINED_DANGER_SCORE.apply(lambda x: 1 if (x>=0 and x<=1) else x)
df_new.COMBINED_DANGER_SCORE = df_new.COMBINED_DANGER_SCORE.apply(lambda x: 2 if (x>1 and x<=2) else x)
df_new.COMBINED_DANGER_SCORE = df_new.COMBINED_DANGER_SCORE.apply(lambda x: 3 if (x>2) else x)
df_new.COMBINED_DANGER_SCORE.unique()
```

Next we further analyzre the data to identify potential uses for it.

```python
environment = ['ANIMAL','EVASIVE ACTION DUE TO ANIMAL, OBJECT, NONMOTORIST','MOTORCYCLE ADVANCING LEGALLY ON RED LIGH',\
              'ROAD CONSTRUCTION/MAINTENANCE','ROAD ENGINEERING/SURFACE/MARKING DEFECTS','VISION OBSCURED (SIGNS, TREE LIMBS, BUILDINGS, ETC.)','WEATHER']
driver_distracted = ['CELL PHONE USE OTHER THAN TEXTING','DISTRACTION - FROM INSIDE VEHICLE','DISTRACTION - FROM OUTSIDE VEHICLE',\
                    'DISTRACTION - OTHER ELECTRONIC DEVICE (NAVIGATION DEVICE, DVD PLAYER, ETC.)',\
                    'OPERATING VEHICLE IN ERRATIC, RECKLESS, CARELESS, NEGLIGENT OR AGGRESSIVE MANNER','TEXTING']
driver_violation = ['DISREGARDING OTHER TRAFFIC SIGNS','DISREGARDING ROAD MARKINGS','DISREGARDING STOP SIGN',\
                   'DISREGARDING TRAFFIC SIGNALS','DISREGARDING YIELD SIGN','DRIVING ON WRONG SIDE/WRONG WAY',\
                   'EXCEEDING AUTHORIZED SPEED LIMIT','EXCEEDING SAFE SPEED FOR CONDITIONS','FAILING TO REDUCE SPEED TO AVOID CRASH','FAILING TO YIELD RIGHT-OF-WAY','FOLLOWING TOO CLOSELY',\
                   'IMPROPER BACKING','IMPROPER LANE USAGE','IMPROPER OVERTAKING/PASSING','IMPROPER TURNING/NO SIGNAL',\
                   'PASSING STOPPED SCHOOL BUS','TURNING RIGHT ON RED']
driver_skill = ['DRIVING SKILLS/KNOWLEDGE/EXPERIENCE']
vehicle_condition =['EQUIPMENT - VEHICLE CONDITION']
driver_condition = ['PHYSICAL CONDITION OF DRIVER','UNDER THE INFLUENCE OF ALCOHOL/DRUGS (USE WHEN ARREST IS EFFECTED)','HAD BEEN DRINKING (USE WHEN ARREST IS NOT MADE)']
unknown = ['NOT APPLICABLE','UNABLE TO DETERMINE']
```

```python
df_new.PRIM_CONTRIBUTORY_CAUSE = df_new.PRIM_CONTRIBUTORY_CAUSE.apply(lambda x: 'environment' if x in environment else x)
df_new.PRIM_CONTRIBUTORY_CAUSE = df_new.PRIM_CONTRIBUTORY_CAUSE.apply(lambda x: 'driver_distracted' if x in driver_distracted else x)
df_new.PRIM_CONTRIBUTORY_CAUSE = df_new.PRIM_CONTRIBUTORY_CAUSE.apply(lambda x: 'driver_violation' if x in driver_violation else x)
df_new.PRIM_CONTRIBUTORY_CAUSE = df_new.PRIM_CONTRIBUTORY_CAUSE.apply(lambda x: 'driver_skill' if x in driver_skill else x)
df_new.PRIM_CONTRIBUTORY_CAUSE = df_new.PRIM_CONTRIBUTORY_CAUSE.apply(lambda x: 'vehicle_condition' if x in vehicle_condition else x)
df_new.PRIM_CONTRIBUTORY_CAUSE = df_new.PRIM_CONTRIBUTORY_CAUSE.apply(lambda x: 'driver_condition' if x in driver_condition else x)
df_new.PRIM_CONTRIBUTORY_CAUSE = df_new.PRIM_CONTRIBUTORY_CAUSE.apply(lambda x: 'unknown' if x in unknown else x)
```
